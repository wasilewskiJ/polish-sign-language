# Klasyfikacja znaków polskiego języka migowego

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from datetime import datetime

from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType
from pyspark.ml.linalg import Vectors, VectorUDT
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer
from pyspark.ml.classification import RandomForestClassifier, LogisticRegression
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.mllib.evaluation import MulticlassMetrics
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
```

## Wczytanie datasetu

```python
def load_split_features_as_numpy(data_dir: Path | str, split: str):
    if isinstance(data_dir, str):
        data_dir = Path(data_dir)

    split_dir = data_dir / split
    if not split_dir.exists():
        raise FileNotFoundError(f"Split directory not found: {split_dir}")

    feature_blocks = []
    label_blocks = []

    for file_path in sorted(split_dir.glob("features_*.npy")):
        label_str = file_path.stem.split("_")[1]
        features = np.load(str(file_path))
        labels = np.load(str(split_dir / f"labels_{label_str}.npy"))

        feature_blocks.append(features)
        label_blocks.append(labels)

    X = np.vstack(feature_blocks) if feature_blocks else np.empty((0, 78))
    y = np.concatenate(label_blocks) if label_blocks else np.empty((0,), dtype=object)

    return X, y


def numpy_to_spark_df(spark: SparkSession, X: np.ndarray, y: np.ndarray):
    rows = [(str(y_i), Vectors.dense(x_i.tolist())) for x_i, y_i in zip(X, y)]

    schema = StructType(
        [
            StructField("label_str", StringType(), nullable=False),
            StructField("features", VectorUDT(), nullable=False),
        ]
    )
    return spark.createDataFrame(rows, schema)


spark = SparkSession.builder.appName("uczenie_maszynowe").getOrCreate()
spark.sparkContext.setLogLevel("ERROR")

data_dir = Path("../backend/translator/data")
if not data_dir.exists():
    raise FileNotFoundError(f"Data directory not found: {data_dir.resolve()}")

X_train, y_train = load_split_features_as_numpy(data_dir, "train")
X_val, y_val = load_split_features_as_numpy(data_dir, "val")
X_test, y_test = load_split_features_as_numpy(data_dir, "test")

print("Train:", X_train.shape, y_train.shape)
print("Val:  ", X_val.shape, y_val.shape)
print("Test: ", X_test.shape, y_test.shape)
```

## Metryki

```python
def collect_pred_labels(df_pred):
    rows = df_pred.select("prediction", "label").rdd.map(
        lambda r: (int(r.prediction), int(r.label))
    ).collect()
    if not rows:
        return np.array([], dtype=int), np.array([], dtype=int)
    y_pred = np.array([r[0] for r in rows], dtype=int)
    y_true = np.array([r[1] for r in rows], dtype=int)
    return y_true, y_pred


def per_class_accuracy(cm, class_labels):
    acc = {}
    for idx, label in enumerate(class_labels):
        row_sum = cm[idx].sum()
        acc[label] = float(cm[idx, idx] / row_sum) if row_sum else 0.0
    return acc


def evaluate_multiclass(df_pred, split_name: str, class_labels):
    evaluator_acc = MulticlassClassificationEvaluator(
        labelCol="label", predictionCol="prediction", metricName="accuracy"
    )
    evaluator_f1w = MulticlassClassificationEvaluator(
        labelCol="label", predictionCol="prediction", metricName="f1"
    )
    evaluator_precision_w = MulticlassClassificationEvaluator(
        labelCol="label", predictionCol="prediction", metricName="weightedPrecision"
    )
    evaluator_recall_w = MulticlassClassificationEvaluator(
        labelCol="label", predictionCol="prediction", metricName="weightedRecall"
    )

    acc = evaluator_acc.evaluate(df_pred)
    f1w = evaluator_f1w.evaluate(df_pred)
    pw = evaluator_precision_w.evaluate(df_pred)
    rw = evaluator_recall_w.evaluate(df_pred)

    y_true, y_pred = collect_pred_labels(df_pred)
    label_ids = list(range(len(class_labels)))
    cm = confusion_matrix(y_true, y_pred, labels=label_ids) if len(y_true) else np.zeros((0, 0))
    per_class = per_class_accuracy(cm, class_labels) if len(y_true) else {}

    print(f"\n[{split_name}]")
    print(f"Accuracy:          {acc:.4f}")
    print(f"Weighted F1:       {f1w:.4f}")
    print(f"Weighted Precision:{pw:.4f}")
    print(f"Weighted Recall:   {rw:.4f}")

    return {
        "accuracy": acc,
        "f1_weighted": f1w,
        "precision_weighted": pw,
        "recall_weighted": rw,
    }, cm, per_class
```

```python
results_dir = Path("results")
results_dir.mkdir(parents=True, exist_ok=True)
results = []

def save_confusion_matrix(cm, class_labels, title: str, path: Path):
    fig, ax = plt.subplots(figsize=(10, 10))
    im = ax.imshow(cm)

    ax.set_title(title)
    ax.set_xlabel("Predicted label")
    ax.set_ylabel("True label")

    ax.set_xticks(range(len(class_labels)))
    ax.set_yticks(range(len(class_labels)))
    ax.set_xticklabels(class_labels, rotation=90)
    ax.set_yticklabels(class_labels)

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            value = cm[i, j]
            if value > 0:
                ax.text(
                    j,
                    i,
                    str(int(value)),
                    ha="center",
                    va="center",
                    color="white" if value > cm.max() * 0.5 else "black",
                    fontsize=8,
                )

    fig.tight_layout()
    fig.savefig(path, dpi=150)
    plt.close(fig)


def write_report(results, output_path: Path):
    lines = []
    lines.append("# Wyniki eksperymentow\n")
    lines.append(f"Data: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")

    for result in results:
        lines.append(f"## {result['model']}\n")
        lines.append("Metryki (test):\n")
        lines.append(f"- Accuracy: {result['metrics']['accuracy']:.4f}\n")
        lines.append(f"- Weighted F1: {result['metrics']['f1_weighted']:.4f}\n")
        lines.append(f"- Weighted Precision: {result['metrics']['precision_weighted']:.4f}\n")
        lines.append(f"- Weighted Recall: {result['metrics']['recall_weighted']:.4f}\n")

        if result.get('confusion_matrix_path'):
            lines.append(f"Macierz pomylek: {result['confusion_matrix_path']}\n")

        if result.get('per_class_accuracy'):
            lines.append("Per-class accuracy:\n")
            lines.append("| Class | Accuracy |\n")
            lines.append("|---|---|\n")
            for label, acc in result['per_class_accuracy'].items():
                lines.append(f"| {label} | {acc:.4f} |\n")

        lines.append("\n")

    output_path.write_text("".join(lines), encoding="utf-8")
```

## Przygotowanie danych (train/val/test)


```python
train_df = numpy_to_spark_df(spark, X_train, y_train)
val_df = numpy_to_spark_df(spark, X_val, y_val)
test_df = numpy_to_spark_df(spark, X_test, y_test)

train_df = train_df.cache()
val_df = val_df.cache()
test_df = test_df.cache()

seed = 10
```

## Klasyfikator RandomForest

```python
label_indexer = StringIndexer(
    inputCol="label_str", outputCol="label", handleInvalid="error"
)

rf = RandomForestClassifier(
    featuresCol="features",
    labelCol="label",
    predictionCol="prediction",
    probabilityCol="probability",
    rawPredictionCol="rawPrediction",
    seed=seed,
)

pipeline = Pipeline(stages=[label_indexer, rf])
param_grid = (
    ParamGridBuilder()
    .addGrid(rf.numTrees, [50, 100, 200])
    .addGrid(rf.maxDepth, [5, 10, 15])
    .addGrid(rf.featureSubsetStrategy, ["sqrt", "log2"])
    .build()
)


evaluator_f1 = MulticlassClassificationEvaluator(
    labelCol="label", predictionCol="prediction", metricName="f1"
)

cv = CrossValidator(
    estimator=pipeline,
    estimatorParamMaps=param_grid,
    evaluator=evaluator_f1,
    numFolds=5,
    seed=seed,
)


cv_model = cv.fit(train_df)
best_model = cv_model.bestModel
```

## Ewaluacja

```python
test_pred = best_model.transform(test_df)
indexer_model = best_model.stages[0]
class_labels = indexer_model.labels

metrics_rf, cm_rf, per_class_rf = evaluate_multiclass(
    test_pred, "TEST - RandomForest", class_labels
)

cm_path = results_dir / "confusion_random_forest.png"
save_confusion_matrix(
    cm_rf, class_labels, "Confusion Matrix (TEST) - RandomForest", cm_path
)

results.append(
    {
        "model": "RandomForest",
        "metrics": metrics_rf,
        "per_class_accuracy": per_class_rf,
        "confusion_matrix_path": str(cm_path),
    }
)
```

## Klasyfikator Logistic Regression

```python
lr = LogisticRegression(
    featuresCol="features",
    labelCol="label",
    predictionCol="prediction",
    probabilityCol="probability",
    rawPredictionCol="rawPrediction",
    maxIter=200,
)

pipeline_lr = Pipeline(stages=[label_indexer, lr])

param_grid_lr = (
    ParamGridBuilder()
    .addGrid(lr.regParam, [0.0, 0.01, 0.1])
    .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0])
    .build()
)

cv_lr = CrossValidator(
    estimator=pipeline_lr,
    estimatorParamMaps=param_grid_lr,
    evaluator=evaluator_f1,
    numFolds=5,
    seed=seed,
)

cv_model_lr = cv_lr.fit(train_df)
best_model_lr = cv_model_lr.bestModel
```

## Ewaluacja

```python
test_pred_lr = best_model_lr.transform(test_df)
indexer_model_lr = best_model_lr.stages[0]
class_labels_lr = indexer_model_lr.labels

metrics_lr, cm_lr, per_class_lr = evaluate_multiclass(
    test_pred_lr, "TEST - LogisticRegression", class_labels_lr
)

cm_path_lr = results_dir / "confusion_logistic_regression.png"
save_confusion_matrix(
    cm_lr, class_labels_lr, "Confusion Matrix (TEST) - LogisticRegression", cm_path_lr
)

results.append(
    {
        "model": "LogisticRegression",
        "metrics": metrics_lr,
        "per_class_accuracy": per_class_lr,
        "confusion_matrix_path": str(cm_path_lr),
    }
)
```

## Custom Keras (MLP) model - load & evaluate\n

```python
tf.random.set_seed(seed)

model_path = Path("../backend/translator/models/pjm_model.keras")
if not model_path.exists():
    raise FileNotFoundError(f"Model not found: {model_path.resolve()}")

keras_model = tf.keras.models.load_model(model_path)

label_encoder = LabelEncoder()
y_train_enc = label_encoder.fit_transform(y_train)
y_val_enc = label_encoder.transform(y_val)
y_test_enc = label_encoder.transform(y_test)

test_probs = keras_model.predict(X_test, verbose=0)
test_pred = np.argmax(test_probs, axis=1)

acc = accuracy_score(y_test_enc, test_pred)
f1w = f1_score(y_test_enc, test_pred, average="weighted")
pw = precision_score(y_test_enc, test_pred, average="weighted")
rw = recall_score(y_test_enc, test_pred, average="weighted")

print("\n[TEST - Keras MLP]")
print(f"Accuracy:          {acc:.4f}")
print(f"Weighted F1:       {f1w:.4f}")
print(f"Weighted Precision:{pw:.4f}")
print(f"Weighted Recall:   {rw:.4f}")

label_ids_keras = list(range(len(label_encoder.classes_)))
cm_keras = confusion_matrix(y_test_enc, test_pred, labels=label_ids_keras)
per_class_keras = per_class_accuracy(cm_keras, label_encoder.classes_)

cm_path_keras = results_dir / "confusion_keras_mlp.png"
save_confusion_matrix(
    cm_keras, label_encoder.classes_, "Confusion Matrix (TEST) - Keras MLP", cm_path_keras
)

results.append(
    {
        "model": "Keras MLP",
        "metrics": {
            "accuracy": acc,
            "f1_weighted": f1w,
            "precision_weighted": pw,
            "recall_weighted": rw,
        },
        "per_class_accuracy": per_class_keras,
        "confusion_matrix_path": str(cm_path_keras),
    }
)
```

```python
report_path = results_dir / "report.md"
write_report(results, report_path)
print(f"Report saved to {report_path.resolve()}")
```
